# Restaurant Management System

```python
class FMS:
    
    def __init__(self,rest_name,rest_menu):
        self.restaurant_name=rest_name
        self.menu=rest_menu
        self.total_bill=0
        self.user_pay=0
        self.user_order=''
        self.welcome_label=(f'Welcome to {self.restaurant_name.title()}!')
        
    def welcome_user(self):
        print(f'Welcome to {self.restaurant_name.title()}!')
        
    def display_menu(self):
        print('*'*51)
        print("Menu:")
        for i in self.menu:
            print(i.title(),self.menu[i])
        print('*'*51)

    def take_order(self):
        self.user_order=input("Please place your order here.")
        
    def preparing_order(self):
        from countdown import countdown 
        print(f"Great, preparing {self.user_order.title()} for you now and it'll be ready in:")
        countdown(mins=0,secs=30)
        self.total_bill=self.total_bill+self.menu[self.user_order.lower()]
        
        
    def serve_order(self):
        print(f"Your {self.user_order.title()} is ready. Enjoy!")
        
    def display_bill(self):
        print(f"Your total bill is {self.total_bill}")
        
    def verify_payment(self):
        self.user_pay=int(input("Please pay your bill here."))
        
        while self.total_bill>self.user_pay:
            self.total_bill = self.total_bill - self.user_pay
            print(f"Thanks. The bill is {self.total_bill} and we've received {self.user_pay}.\n Please pay additional ${self.total_bill}.")
            self.user_pay=int(input(f"Please pay additional ${self.total_bill} here."))
        if self.user_pay>self.total_bill:
            print(f"Thanks! Here's your change: ${self.user_pay-self.total_bill}")
        else:
            pass
        
    def order_process(self):
        self.take_order()
        if self.user_order.lower() in self.menu:
            self.preparing_order()
            self.serve_order()
            self.user_repeat_order=input("Would you like to order anything else?")
            while self.user_repeat_order.lower()=='yes':
                self.repeat_order()
                self.user_repeat_order=input("Would you like to order anything else?")
                
            self.display_bill()
            self.verify_payment()
            self.thank_user()
            
        else:
            print(f"Sorry, {self.user_order.title()} is not available. Please place order from the menu.")
            self.order_process()
                  
    def repeat_order(self):
        self.take_order()
        if self.user_order.lower() in self.menu:
            self.preparing_order()
            self.serve_order()
        else:
            print(f"Sorry, {self.user_order.title()} is not available. Please place order from the menu.")
            self.repeat_order()
            
    def thank_user(self):
        print(f"Hope you enjoyed the beverage(s). Visit again....")
            
if __name__ == '__main__':
    
    file = open('user input.txt')
    
    user_input=file.readlines()
    
    rn=user_input[0].replace('\n','')
    
    food_name = user_input[1].replace('\n','').split(',')
    
    food_prices=[]
    for i in user_input[2].split(','):
        food_prices.append(int(i))
        
    rm=dict(zip(food_name,food_prices))
    
    dummy_rest=(FMS(rest_name=rn,rest_menu=rm))
    
    
    import tkinter as tk
    window=tk.Tk()
    window.title('FMS')
    window.geometry('300x300')
    tk.Label(window,text=dummy_rest.welcome_label,font=('Calibri',16)).place(x=30,y=10)
    tk.Button(window,text='Menu',command=dummy_rest.display_menu).place(x=120,y=80)
    tk.Button(window,text='Start Ordering',command=dummy_rest.order_process).place(x=100,y=160)
    
    window.mainloop()
```
